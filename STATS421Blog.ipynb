# Cilppers and Lakers Salary Analysis

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
# Example data structure
data = pd.read_excel("NBAClipperLaker.xlsx")
# Drop any rows with missing values (if applicable)
data = data.dropna()
data
```

```python
data.sort_values("Salary")
```

```python
# Create the box plot for Salary
plt.figure(figsize=(8, 6))
plt.boxplot(data['Salary'], vert=False)
plt.title('Box Plot of NBA Player Salaries')
plt.xlabel('Salary ($)')
plt.show()
```

```python
data.sort_values("PTS")
```

```python
data.dtypes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only the relevant columns (Salary, PTS, MIN) for the correlation matrix
correlation_data = data[["Salary", "PTS", "MIN"]]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Plot the correlation matrix heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu", fmt=".2f", square=True, cbar_kws={"shrink": .8})
plt.title("Correlation Matrix Heatmap for Salary, PTS, and MIN")
plt.show()
```

```python
# Scatter plot with regression planes
plt.figure(figsize=(12, 6))

# Salary vs. PTS
plt.subplot(1, 2, 1)
sns.scatterplot(x=data["PTS"], y=data["Salary"], color="blue", alpha=0.5)
plt.xlabel("PTS")
plt.ylabel("Salary")
plt.title("Salary vs. PTS")

# Salary vs. MINS
plt.subplot(1, 2, 2)
sns.scatterplot(x=data["MIN"], y=data["Salary"], color="green", alpha=0.5)
plt.xlabel("MINS")
plt.ylabel("Salary")
plt.title("Salary vs. MINS")

plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x=data["MIN"], y=data["PTS"], color="orange", alpha=0.5)
plt.xlabel("MIN")
plt.ylabel("PTS")
plt.title("MIN vs. PTS")
```

```python
# Define the independent variables (PTS and MINS) and the dependent variable (Salary)
X = data[["PTS", "MIN"]]  # Independent variables
y = data["Salary"]         # Dependent variable
```

```python
# Add a constant to the model for the intercept
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Define the independent variables (PTS and MINS) and the dependent variable (Salary)
X = data[["PTS"]]  # Independent variables
y = data["Salary"]         # Dependent variable
# Add a constant to the model for the intercept
X = sm.add_constant(X)
# Fit the OLS model
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data["PTS"], y=data["Salary"], color="blue", alpha=0.5)
plt.plot(data["PTS"], model.predict(X), color="red", linewidth=2, label="Regression Line")
plt.xlabel("PTS")
plt.ylabel("Salary")
plt.title("Scatter Plot of Salary vs. PTS with Regression Line")
plt.legend()
plt.show()
```
